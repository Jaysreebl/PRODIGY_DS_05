```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
num_cities = 150
cities = [f"City {i}" for i in range(1, num_cities + 1)]
road_conditions = ['dry', 'wet', 'icy']
weather_conditions = ['clear', 'rainy', 'snowy']
times_of_day = ['morning', 'afternoon', 'evening']
accident_severity = ['minor', 'moderate', 'severe']

np.random.seed(0)
data = {
    'City': np.random.choice(cities, size=num_cities),
    'Road_Conditions': np.random.choice(road_conditions, size=num_cities),
    'Weather_Conditions': np.random.choice(weather_conditions, size=num_cities),
    'Time_of_Day': np.random.choice(times_of_day, size=num_cities),
    'Accident_Severity': np.random.choice(accident_severity, size=num_cities)
}
df = pd.DataFrame(data)
print("Data Exploration and Cleaning:")
print(df.info())
```

```python
# Road Conditions and Weather
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Road_Conditions', palette='viridis', hue='Road_Conditions', legend=False)
plt.title('Distribution of Road Conditions')

plt.subplot(1, 2, 2)
sns.countplot(data=df, x='Weather_Conditions', palette='viridis', hue='Weather_Conditions', legend=False)
plt.title('Distribution of Weather Conditions')

plt.tight_layout()
plt.show()
```

```python
# Time of the Day
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Time_of_Day', palette='viridis', hue='Weather_Conditions', legend=False)
plt.title('Distribution of Time of Day')
plt.show()
```

```python
# Identifying Accident Hotspots (simulated geographical plot)
plt.figure(figsize=(10, 8))
plt.scatter(np.random.rand(num_cities), np.random.rand(num_cities), alpha=0.5)
plt.title('Simulated Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Contributing Factors
pivot_table = pd.pivot_table(df, index='Road_Conditions', columns='Accident_Severity', aggfunc='size', fill_value=0)
print("\nAccident Severity by Road Conditions:")
print(pivot_table)
```

```python
# Example: Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm')
plt.title('Accident Severity by Road Conditions')
plt.show()
```
